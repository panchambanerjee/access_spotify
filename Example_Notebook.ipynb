### Import the relevant modules

```python
from access_spotify.get_spotify_access import get_spotify_credentials
from access_spotify.get_album_details import get_album_details
from access_spotify.get_album_art import get_album_art
from access_spotify.get_album_data import get_album_audio_analysis, get_album_audio_features
```

### Get the Client ID and Client Secret for accessing the Spotify API

* Get your Spotify access credentials here: https://developer.spotify.com/dashboard/login
* Then populate the spotify_credentials.py with your data

```python
from spotify_credentials import CLIENT_ID, CLIENT_SECRET

sp = get_spotify_credentials(client_id=CLIENT_ID, client_secret=CLIENT_SECRET)
```

### Specify the artist and the album you want to query for (note: running the script will download all the album art and data for an artist). 

For instructions on how to run the script, please see the README. 

```python
artist_name = 'Radiohead'
album_names, album_name_uri_dict, album_img_url_dict = get_album_details(sp=sp, artist_name=artist_name)
album_names
```

```python
# Specify which album you want to get the album art and data for
album_name = 'In Rainbows'
```

### Get the album art (in high resolution) 

```python
get_album_art(album_name=album_name, album_img_url_dict=album_img_url_dict,
                      album_cover_path='./')
```

### Get the audio features and audio analysis for all tracks on the album

```python
df_audio_features = get_album_audio_features(sp=sp, album_name=album_name, album_name_dict=album_name_uri_dict,
                                 album_info_path='./')
df_audio_features
```

```python
df_audio_analysis = get_album_audio_analysis(sp=sp, album_name=album_name, album_name_dict=album_name_uri_dict,
                                 album_info_path='./')
df_audio_analysis
```
